```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

```

```python
df = pd.read_csv('titanic.csv')
```

```python
print(df)
```

```python
df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

print(df)
```

```python
df = df[[c for c in df if c not in ['Survived']] 
       + ['Survived']]
print(df)
```

```python
print(df['Sex'])
arr = []
for a in df['Sex'] :
    if a == 'male': arr.append(1)
    if a == 'female': arr.append(0)
print(arr)
```

```python
df.drop(['Sex'], axis=1, inplace=True)
# print(df['Sex'])
df.insert(1,'Sex',arr)
print(df['Sex'])
```

```python
print(df[0:5])
```

```python
print(df.dtypes)
```

```python
b = df['Survived'].value_counts()
print(b)
# ax = df.plot.bar(x='Survived')
plt.bar(x=[0,0.4], height=b, width=0.35)
plt.xticks([0,.4], ("Perished", "Survived"))

plt.ylabel('Outcomes')
plt.title('Titanic Survival Numbers')
plt.show()
```

```python
df.mean()
```

```python
df.groupby(df['Survived']).mean()
```

```python
df.groupby(df['Survived']).mean().iloc[0]
```

```python
samples = np.arange(1,7)
plt.close()
plt.plot(samples, df.groupby(df['Survived']).mean().iloc[0],label="Perished")
plt.plot(samples, df.groupby(df['Survived']).mean().iloc[1],label="Survived")
plt.plot(samples, df.mean().drop('Survived'),label="Total Mean")
plt.title("Feature Averages")
plt.xticks(samples, ["Pclass","Sex","Age","Number Siblings","Number children/parents","Fare"])
plt.xlabel("Feature")
plt.ylabel("Average")
plt.legend()
```

```python
count = 0
for a in df.drop(['Survived','Embarked'], axis=1).columns.values:
    arr = df.mean().iloc[count]
#     print(arr)
    b = df.groupby(df['Survived']).mean()[a]
    b=b.to_frame()
    b.loc[2] = arr
    b=b.squeeze()
#     print(b)
    # ax = df.plot.bar(x='Survived')
    plt.bar(x=[0,1,2], height=b, width=0.35)
    plt.xticks([0,1,2], ("Perished", "Survived", "Average"))
    
#     plt.ylabel(a)
    plt.title(a)
    plt.show()
    count = count + 1
```

```python
for a in df.drop(['Survived','Embarked'], axis=1):
#     print(a)
    print("Correlation between Survived and",a,"is",df['Survived'].corr(df[a]))
```
